```python
import cv2
import numpy as np
import time
import math
```

```python
cap = cv2.VideoCapture("main video.mp4")
fps =25
while cap.isOpened():
    ret,frame = cap.read()
    if ret:
        n1_frame=node1(frame)
        n2_frame=node2(n1_frame)
        n3_frame=node3(n2_frame,frame)
        time.sleep(1/fps)
        cv2.imshow("task video" , n3_frame)
    
        if cv2.waitKey(20) == 27:
            break
    else:
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
def node1(frame):
    blur = cv2.GaussianBlur(frame,(9,9),0)
    hsv = cv2.cvtColor(blur,cv2.COLOR_BGR2HSV)
    low_range = np.array([20,80,130])
    high_range = np.array([39,255,255])
    mask = cv2.inRange(hsv,low_range,high_range)
    return mask
```

```python
def node2(frame):
    edge=cv2.Canny(frame,75,150)
    return edge
```

```python
def node3(frame,original):
    lines=cv2.HoughLinesP(frame,1,np.pi/180,40,minLineLength=10, maxLineGap=30)
    if lines is not None:
        x=0
        y=0
        thetas=[]
        for line in lines:
            x1,y1,x2,y2 =line[0]
            x+=x1+x2
            y+=y1+y2
            if(x1-x2==0):
                th = np.pi/2
            else:
                th = math.atan((y2-y1)/(x2-x1))*(180/np.pi)
                #print(th)################3
            thetas.append(th)
            cv2.line(original,(x1,y1),(x2,y2),(0,0,255),5)
        M=max(thetas)
        m=min(thetas)
        #print(f"M={M} m={m}")#################
        x=int(0.5*x/len(thetas))
        y=int(0.5*y/len(thetas))
        #print(f"X={x} Y={y}")##################
        if (M-m)>1:
            theta = round(180-(M-m),2)
            cv2.putText(original,"angle="+str(theta),(x,y+2),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,255),2,cv2.LINE_AA)
        else:
            cv2.putText(original,"not found",(x,y+2),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2,cv2.LINE_AA)
    return original
```
